```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
from keras.models import Sequential
from keras.layers.core import Dense, Activation
from keras.utils import np_utils
```

```python
credit_default=pd.read_csv("cleaned_cerditcard.csv")
```

```python
credit_default.head()
```

```python
credit_default.shape
```

```python
credit_default = credit_default.drop(credit_default.columns[[0]], axis=1)
```

```python
credit_default.head()
```

```python
X= credit_default.drop('default', axis=1)
```

```python
X.head()
```

```python
traget_name = 'default'
y= credit_default[traget_name]
```

```python
y.head()
```

```python
train_data_x, test_data_x, train_data_y, test_data_y =\
train_test_split(X,y, train_size=0.5, test_size=0.5, random_state=0)
```

```python
train_data_x.head()
```

```python
train_data_y.head()
```

```python
model = Sequential()
```

```python
model.add(Dense(16, input_shape=(25,), name='Input_Layer', activation='relu'))
```

```python
# 16 neurons, 
# 4 features in the dataset
```

```python
# Output Layer
model.add(Dense(2, name='Output_Layer', activation='softmax'))
```

```python
# 3 neuron b/c we want to classify the output or flower into 3 categories.
```

```python
model.summary()
```

```python
# compile
model.compile(optimizer='sgd', loss='sparse_categorical_crossentropy', metrics=["accuracy"])
```

```python
# adam => Adaptive momentum-based optimizer-this generally works well for deep NNs
# crossentropy => mean that the output is the probability distibution
# low crossentropy mean y_actual == y_predicted.
```

```python
# Start training
model.fit(train_data_x, train_data_y, epochs=50, batch_size=2, verbose=1)
```

```python
loss, accuracy = model.evaluate(test_data_x, test_data_y, verbose=0)
```

```python
print("Loss={:.2f}".format(loss))
```

```python
print("Accuracy = {:.2f}".format(accuracy))
```

```python
input_dim = train_data_x.shape[1]

model = Sequential()
model.add(Dense(256, input_shape=(input_dim,), activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(1,  activation='sigmoid'))
```

```python
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(
              np.array(train_data_x), np.array(train_data_y),
              batch_size=25, epochs=50, verbose=1,
              validation_data=(np.array(test_data_x), np.array(test_data_y)))
```

```python
score = model.evaluate(np.array(test_data_x), np.array(test_data_y), verbose=0)
print('Test Loss:', score[0])
print('Test Accuracy:', score[1])
```

```python
model2 = Sequential()
model2.add(Dense(256, input_shape=(input_dim,), activation='relu', weights=model.layers[0].get_weights()))
model2.add(Dense(256, activation='relu', weights=model.layers[1].get_weights()))
model2.add(Dense(64, activation='relu', weights=model.layers[2].get_weights()))
model2.add(Dense(64, activation='relu', weights=model.layers[3].get_weights()))
model2.add(Dense(10, activation='sigmoid', weights=model.layers[4].get_weights()))
```

```python
activations = model2.predict_on_batch(np.array(test_data_x))
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, perplexity=25, verbose=0, n_iter=500, random_state=1337)
samples = tsne.fit_transform(activations)
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots(1,2)
fig.set_size_inches(20,10)

ax[0].scatter(*samples[test_data_y==0].T,color='b', alpha=0.5, label='default: NO')
ax[1].scatter(*samples[test_data_y==1].T,color='r', alpha=0.5, label='default: YES')

plt.tight_layout()
plt.show()
```
