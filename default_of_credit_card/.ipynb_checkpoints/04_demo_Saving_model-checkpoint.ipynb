```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
from keras.models import Sequential
from keras.layers.core import Dense, Activation
from keras.utils import np_utils
```

```python
credit_default=pd.read_csv("cleaned_cerditcard.csv")
```

```python
credit_default.head()
```

```python
credit_default = credit_default.drop(credit_default.columns[[0]], axis=1)
```

```python
credit_default.head()
```

```python
credit_default.columns
```

```python
credit_default.shape
```

```python
X= credit_default.drop('default', axis=1)
```

```python
X.head()
```

```python
traget_name = 'default'
```

```python
y= credit_default[traget_name]
```

```python
y
```

```python
train_data_x, test_data_x, train_data_y, test_data_y =\
train_test_split(X,y, train_size=0.5, test_size=0.5, random_state=0)
```

```python
train_data_x.columns
```

```python
model = Sequential()
```

```python
model.add(Dense(30, input_shape=(25,), name='Input_Layer', activation='relu'))
```

```python
# 30 neurons, 
# 25 features in the dataset
```

```python
# Output Layer
model.add(Dense(2, name='Output_Layer', activation='softmax'))
```

```python
# 2 neuron b/c we want to classify the output or nature of payment into 2 categories.
```

```python
model.summary()
```

```python
# compile
model.compile(optimizer='sgd', loss='sparse_categorical_crossentropy', metrics=["accuracy"])
```

```python
# adam => Adaptive momentum-based optimizer-this generally works well for deep NNs
# crossentropy => mean that the output is the probability distibution
# low crossentropy mean y_actual == y_predicted.
```

```python
# Start training
model.fit(train_data_x, train_data_y, epochs=80, batch_size=2, verbose=1)
```

```python
loss, accuracy = model.evaluate(test_data_x, test_data_y, verbose=0)
```

```python
print("Loss={:.2f}".format(loss))
```

```python
print("Accuracy = {:.2f}".format(accuracy))
```
